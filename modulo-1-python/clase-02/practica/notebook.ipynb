### Ejercicio 5 — Reporte por categoría

En un reporte de ventas real necesitas agrupar los datos por algún criterio — categoría, región, vendedor — y calcular métricas por grupo. Antes de usar pandas, este ejercicio te muestra cómo funciona esa lógica por dentro.

**Consigna:** Recorre la lista de productos y construí un diccionario que agrupe el total de ventas (precio × cantidad) por categoría. Al final imprimí un reporte con el total por categoría y cuál fue la categoría con mayor ingreso.

```python
# Punto de partida
productos = [
    {"nombre": "Notebook",    "categoria": "tecnologia", "precio": 320000, "cantidad": 3},
    {"nombre": "Escritorio",  "categoria": "muebles",    "precio": 85000,  "cantidad": 5},
    {"nombre": "Mouse",       "categoria": "tecnologia", "precio": 8500,   "cantidad": 20},
    {"nombre": "Silla",       "categoria": "muebles",    "precio": 45000,  "cantidad": 8},
    {"nombre": "Monitor",     "categoria": "tecnologia", "precio": 210000, "cantidad": 4},
    {"nombre": "Lámpara",     "categoria": "muebles",    "precio": 12000,  "cantidad": 15},
    {"nombre": "Auriculares", "categoria": "tecnologia", "precio": 32000,  "cantidad": 11},
]

totales_por_categoria = {}

# Tu tarea: construir el diccionario de totales y encontrar la categoría ganadora

tech = []
hogar = []

for prod in productos:
    cat = prod["categoria"]
    if cat == "muebles":
        hogar.append(prod["precio"] * prod["cantidad"])
    elif cat == "tecnologia":
        tech.append(prod["precio"] * prod["cantidad"])

total_ventas_tech = sum(tech)
total_ventas_hogar = sum(hogar)

if total_ventas_tech < total_ventas_hogar:
    categoria_ganadora = "hogar"
else: categoria_ganadora = "tecnologia"


reporte = f" Las ventas finales para tecnologia suman un total de: ${total_ventas_tech} \n Las ventas finales para hogar suman un total de: ${total_ventas_hogar} \n La categoria ganadora es: {categoria_ganadora}."

print(reporte)
```

En la práctica, los datos que llegan de fuentes externas rara vez están limpios. Valores negativos, campos vacíos, tipos incorrectos — todo eso aparece antes de poder analizar nada. Saber detectarlo y manejarlo es una habilidad central en data analysis.

**Consigna:** Recorre la lista de registros e identificá los que tienen problemas. Un registro es inválido si el precio es negativo, si el nombre está vacío, o si el stock no es un número entero. Separalos en dos listas: validos e invalidos, y para cada inválido indicá el motivo.

```python
# Punto de partida
registros = [
    {"nombre": "Notebook",   "precio": 320000, "stock": 5},
    {"nombre": "",           "precio": 8500,   "stock": 10},
    {"nombre": "Monitor",    "precio": -210000,"stock": 2},
    {"nombre": "Teclado",    "precio": 45000,  "stock": 3.5},
    {"nombre": "Mouse",      "precio": 8500,   "stock": 0},
    {"nombre": "Auriculares","precio": 32000,  "stock": 7},
    {"nombre": "Cámara",     "precio": 0,      "stock": 4},
]

validos = []
invalidos = []

# Tu tarea: clasificar cada registro y registrar el motivo de rechazo
# Ejemplo de salida para inválidos:
# {"nombre": "", "precio": 8500, "stock": 10, "motivo": "nombre vacío"}

def validacion():
    for registro in registros:
        if registro["precio"] < 0:
            invalidos.append(registro)
            print(f"Precio inválido: Debe ser positivo. Revisar: {registro}")
        elif isinstance(registro["stock"], (float)):
            invalidos.append(registro)
            print(f"Stock inválido: Debe ser un valor entero. Revisar: {registro}")
        elif registro["nombre"] == "":
            invalidos.append(registro)
            print(f"Nombre inválido: Inexistente. Revisar: {registro}")
        else: validos.append(registro)

validacion()
```

### Ejercicio 7 — Generador de reportes 

En un sistema real de ventas, los reportes se generan automáticamente a partir de los datos crudos. Este ejercicio simula exactamente eso: a partir de una lista de transacciones, construís un reporte completo que podría enviarse por email o guardarse en un archivo.

**Consigna:** A partir de la lista de transacciones, construye un reporte que incluya:


- Total de transacciones procesadas
- Monto total y promedio
- Desglose por país (cantidad y monto total)
- Las 3 transacciones de mayor monto
- Transacciones rechazadas (monto negativo o cero) con su motivo

```python
# Punto de partida
transacciones = [
    {"id": "TX001", "pais": "MX", "monto": 45000,  "moneda": "MXN"},
    {"id": "TX002", "pais": "AR", "monto": 128000, "moneda": "ARS"},
    {"id": "TX003", "pais": "CO", "monto": 0,      "moneda": "COP"},
    {"id": "TX004", "pais": "MX", "monto": 87000,  "moneda": "MXN"},
    {"id": "TX005", "pais": "AR", "monto": -5000,  "moneda": "ARS"},
    {"id": "TX006", "pais": "PE", "monto": 32000,  "moneda": "PEN"},
    {"id": "TX007", "pais": "CO", "monto": 215000, "moneda": "COP"},
    {"id": "TX008", "pais": "MX", "monto": 61000,  "moneda": "MXN"},
    {"id": "TX009", "pais": "AR", "monto": 94000,  "moneda": "ARS"},
    {"id": "TX010", "pais": "PE", "monto": 17000,  "moneda": "PEN"},
]

transacciones_rechazadas = []
transacciones_AR = []
transacciones_PE = []
transacciones_MX = []
transacciones_CO = []
transacciones_sin_catalogar = []
transacciones_promedio_total = sum(transaccion["monto"] for transaccion in transacciones) / len(transacciones)
mejores_3 = sorted((transaccion["monto"] for transaccion in transacciones), reverse = True)[:3]

for transaccion in transacciones:
    if transaccion["pais"] == "AR":
        transacciones_AR.append(transaccion)
    elif transaccion["pais"] == "CO":
        transacciones_CO.append(transaccion)
    elif transaccion["pais"] == "MX":
        transacciones_MX.append(transaccion)
    elif transaccion["pais"] == "PE":
        transacciones_PE.append(transaccion)
    else: transacciones_sin_catalogar.append(transaccion)
for transaccion in transacciones:
    if transaccion["monto"] < 0:
        transacciones_rechazadas.append(transaccion)

reporte = f"Cantidad de transacciones: {len(transacciones)} \n Monto total de transacciones: ${sum(transaccion["monto"] for transaccion in transacciones):.2f} \n Monto promedio por transacción: ${transacciones_promedio_total:.2f} \n Métricas por país: \n AR: Cantidad: {len(transacciones_AR)} Promedio: ${sum(transaccion_AR["monto"] for transaccion_AR in transacciones_AR) / len(transacciones_AR):.2f} \n MX: Cantidad: {len(transacciones_MX)} Promedio: ${sum(transaccion_MX["monto"]for transaccion_MX in transacciones_MX) / len(transacciones_MX):.2f} \n CO: Cantidad: {len(transacciones_CO)} Promedio: ${sum(transaccion_CO["monto"] for transaccion_CO in transacciones_CO) / len(transacciones_CO):.2f} \n PE: Cantidad: {len(transacciones_PE)} Promedio: ${sum(transaccion_PE["monto"] for transaccion_PE in transacciones_PE) / len(transacciones_PE):.2f} \n Top 3 mejores ventas: {mejores_3}  \n Transacciones rechazadas: {transacciones_rechazadas}"

print(reporte)

# Tu tarea: construir el reporte completo
# Pista: para las top 3 investigá sorted() con el parámetro key=

```
