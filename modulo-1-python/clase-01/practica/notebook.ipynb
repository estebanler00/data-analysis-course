## Ejercicio 3 — Casting para el análisis

Muchos sistemas de scraping o formularios web devuelven todos los datos como strings, incluso los números. Antes de poder operar con esos datos, tiens que convertirlos al tipo correcto.

**Consigna:** El sistema te entregó estos datos de un producto como strings. Conviertelos al tipo adecuado para poder calcular el precio con descuento y el IVA.



```python
# Datos tal como llegan del sistema externo (todo string)
precio_str = "15990"
descuento_str = "0.15"
cantidad_str = "4"
tiene_iva_str = "True"

# Tu tarea:
# 1. Convierte cada variable al tipo correcto (int, float, bool)
# 2. Calcula el precio final con descuento
# 3. Si tiene IVA, suma un 21% adicional
# 4. Imprime el total para la cantidad indicada

precio = int(precio_str)
descuento = float(descuento_str)
cantidad = int(cantidad_str)
tiene_iva = bool(tiene_iva_str)

precio_final = precio - (precio*descuento)

if tiene_iva_str == "True":
    precio_final = precio - (precio*descuento) + (precio*0.21)
else: precio_final

print("El precio total es de $" + str(precio_final*cantidad))
```

## Ejercicio 4 — Reporte de ventas básico

Tu jefa te pide que armes un mini-reporte de ventas del día para enviarlo por mensaje. Tiene que ser legible y estar en una sola variable.

**Consigna:** Usa las variables dadas para construir un reporte en varias líneas usando \n dentro de un string, o un f-string multilínea con """. El reporte debe incluir todos los campos y calcular el promedio de ventas.

```python
fecha = "2024-11-14"
vendedor = "Carlos Méndez"
ventas = [45000, 32000, 67500, 28000, 91000]
producto_estrella = "Auriculares Bluetooth"

# Tu tarea: construir este reporte
# Reporte de ventas — 2024-11-14
# Vendedor: Carlos Méndez
# Total ventas: $263500
# Promedio por venta: $52700.0
# Producto estrella: Auriculares Bluetooth

#total_ventas podria haberse resuelto con sum(ventas)
total_ventas = ventas[0] + ventas[1] + ventas[2] + ventas[3] + ventas[4]
promedio_ventas = total_ventas / len(ventas)
reporte = f"Reporte de ventas {str(fecha)} \nVendedor: {vendedor} \nTotal ventas: {str(total_ventas)} \nPromedio por venta: {str(promedio_ventas)} \nProducto estrella: {producto_estrella}"

print(reporte)
```
