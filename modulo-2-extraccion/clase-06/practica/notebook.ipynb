# Clase 06 — Introducción a SQL



Para esta clase, la práctica no está en este repositorio — la vas a realizar directamente en **[SQLBolt](https://sqlbolt.com/)**, un sitio con ejercicios interactivos de SQL que corrige tus consultas al instante, sobre una base de datos real con información de películas de Pixar.

No es necesario instalar nada: ejecuta las consultas directamente en el navegador y verás el resultado inmediatamente.

### Lecciones a resolver

Para esta clase, completa las siguientes lecciones en orden. Cada una tiene una breve explicación al inicio y una serie de tareas — resuelve las tareas escribiendo la consulta en el editor que aparece en la página:

| Lección | Tema | Se relaciona con |
|---------|------|-------------------|
| [Lección 1 — SELECT queries 101](https://sqlbolt.com/lesson/select_queries_introduction) | Seleccionar columnas | `SELECT`, `FROM` |
| [Lección 2 — Queries with constraints (Pt. 1)](https://sqlbolt.com/lesson/select_queries_with_constraints) | Filtrar con condiciones numéricas | `WHERE` |
| [Lección 3 — Queries with constraints (Pt. 2)](https://sqlbolt.com/lesson/select_queries_with_constraints_pt_2) | Filtrar con condiciones de texto | `WHERE`, `LIKE` |
| [Lección 6 — Multi-table queries with JOINs](https://sqlbolt.com/lesson/select_queries_with_joins) | Unir dos tablas relacionadas | `JOIN` |

> **Por qué se omiten las Lecciones 4 y 5:** introducen `ORDER BY` y subconsultas, que todavía no se han cubierto en clase. Si te interesa adelantarte, no hay ningún problema en resolverlas también — son una progresión natural y las necesitarás pronto.

### Cómo trabajar cada lección

1. Lee la explicación de la lección — generalmente incluye un ejemplo de la sintaxis antes de las tareas.
2. Observa el esquema de la tabla que aparece (qué columnas tiene, qué tipo de dato guarda cada una).
3. Escribe tu consulta en el editor y ejecútala para ver el resultado.
4. Cada lección tiene varias tareas — debes resolverlas todas para que el sitio te permita avanzar a la siguiente lección.
5. Si te quedas atascado, cada tarea tiene un enlace de **"Stuck? Read this task's Solution"** — úsalo como último recurso, después de haber intentado realmente.

#### Después de resolverlas

→ Cuando termines las cuatro lecciones, puedes revisar tus respuestas contra la [solución comentada de SQLBolt](./solucion.md), que incluye además la explicación del *por qué* de cada consulta, no solo el código.
